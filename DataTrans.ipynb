```python
import pandas as pd
df=pd.read_csv('weather.csv')
```

```python
df.head()
```

```python
df.melt(id_vars=['Date', 'City'], value_vars=['Temperature_C', 'Humidity_%'])
# This is used to reshape a DataFrame from a wide format to a long format.
```

```python
df = df.drop_duplicates(subset=['Date', 'Remarks'])
df.pivot(index='Date', columns='City', values='Temperature_C')

#firstly needed to remove the duplicates because there are duplicate values 
# for the same combination of index and columns.
#pivot function allows to look at the data with different point of view
```

```python
df_exploded = df.explode('Temperature_C')
print(df_exploded)
```
