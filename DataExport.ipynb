```python
import pandas as pd
df=pd.read_csv('weather.csv')
```

```python
df1=df[['City','Remarks','Weather_Condition','Date']]

df2=df[['Temperature_C','Humidity_%','Date']]

df3=pd.concat([df1,df2],axis=1)
```

```python
df3.to_csv('file.csv', index=False)
# concatinated 2 dfs and then stored the df3 into a csv file.
```

```python
df3.to_excel('file.xlsx', index=False)
#This will save the df3 into a excel file
```

```python
pd.read_excel('file.xlsx')
#we can read the excel file using this command
```
