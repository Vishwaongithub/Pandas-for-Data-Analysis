```python
import pandas as pd

df= pd.read_csv('weather.csv')
```

```python
df.head()
```

```python
df1=df[['City','Remarks','Weather_Condition','Date']]

df2=df[['Temperature_C','Humidity_%','Date']]
```

```python
pd.concat([df1,df2],axis=1)
# Using concat command we can concat 2 df's. Axis is default 0 so if you want like this output then 
# set the axis to 1 otherwise leave it to default   
```

```python
pd.merge(df1,df2, on='Date',how='inner')
# Merge() function acts as SQL join for data frames , we join the df's on the common key column.
```
