```python
import pandas as pd
df=pd.read_csv('weather.csv')
```

```python
df.memory_usage(deep=True)
#This is used to know the memory usage off the data frame
```

```python
df.head()
```

```python
df=df.astype({'City':'category'})
print(df.dtypes)
#Converts columns to category to save memory
```

```python
for index, row in df.iterrows():
    print(f"Row {index}: {row['City']} has temperature {row['Temperature_C']}¬∞C with {row['Weather_Condition']}.")

# allow you to iterate over rows in a Pandas DataFrame but is slower and more felxible
```

```python
for row in df.itertuples(index=False):
    print(f"{row.City} -{row.Temperature_C}¬∞C -{row.Weather_Condition}")
    
#allow you to iterate over rows in a Pandas DataFrame , is faster and more efficient
```

```python
df.query('Temperature_C > 20')
#The .query() method in pandas allows filtering DataFrame rows based on a condition using a string expression.
```
