```python
import pandas as pd

df=pd.read_csv("weather.csv")
df
```

```python
df[['City','Temperature_C']]
#You can select columns according to your need and that too multiple
```

```python
df.loc[df['Weather_Condition']=='Rainy']
#loc function in pandas is a very powerful and useful function,
#we can filter, update, and can even work with conditions.
```

```python
df.iloc[2]
#this iloc() function uses integer based indexing   
```

```python
df[df['Weather_Condition']=='Rainy']
#This way you can filter the things that you need based on the conditions
```

```python
df.query('Weather_Condition == "Rainy"')
#this function can also be used to filter out the data but in a SQL- like syntax
```
