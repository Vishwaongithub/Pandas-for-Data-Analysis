```python
import pandas as pd

df=pd.read_csv('weather.csv')
df
```

```python
df.isnull().sum()
#provide us with the count of null values in the dataset for all the columns
```

```python
df['Temperature_C']=df['Temperature_C'].fillna(df['Temperature_C'].mean())
#Here we had null values in the "Temperature_C" column so we filled the missing vlaues with the mean,
#you can also fill it with other methods like forward fill (ffill) and backward fill(bfill)
#to know about them more visit the pandas fillna documentation
```

```python
df.dropna()
#Thsi will drop or remove all the rows/columns with missing values
```

```python
df.duplicated()
#This will identify the duplicates in the dataset
```

```python
df.drop_duplicates()
#This removes all the rows with the duplicate values
```

```python
df.drop_duplicates()
#This removes all the rows with the duplicate values
```

```python
df['Temperature_C'].dtype
```

```python
df['Temperature_C']=df['Temperature_C'].astype(dtype=int)
df['Temperature_C'].dtype
#This astype() function changes the data type of a column, as you can see above the data type of 
# column temperature was float but after performing this the data type of column temperature became int
```

```python
pd.to_datetime(df['Date'])
#This changes the column to datetime format
```

```python
df.replace('New York','N.Y')
#This function replaces the values in the columns as here i replaced New York with N.Y
```

```python
df.rename(columns={'Temperature_C':'Temp_C'})
#Changes the column name, Temperature_C changed to Temp_C
```

```python
def mean_wind(x):
    if pd.isna(x):
        return df['Wind_Speed_kmph'].mean()
    else:
        return x
df['Wind_Speed_kmph']=df['Wind_Speed_kmph'].apply(func=mean_wind)
df

# In this the apply() function is used to apply function to a column in the data frame
#here I created a function which will replace all the NaN values with the mean for the wind_speed_kmph column.
```
